```python
lob = lob.dropna().reset_index(drop=True)
# Extract `bid1` and `ask1`
lob['bid1'] = lob['bids'].apply(lambda x: float(x.split('|')[0].split(',')[0]) if pd.notna(x) else None)
lob['ask1'] = lob['asks'].apply(lambda x: float(x.split('|')[0].split(',')[0]) if pd.notna(x) else None)
```

```python
# Midpoint Price
lob['MP'] = (lob['bid1'] + lob['ask1']) / 2
```

```python
# Midpoint Price Change Rate
lob['datetime'] = pd.to_datetime(lob['datetime'])
# Define a function to get the midpoint prices at `t-k`
def get_closest_time(lob, current_time, delta):
    target_time = current_time - delta
    #逻辑是否需要换？
    closest_time = lob['datetime'][lob['datetime'] <= target_time].max()
    if pd.isna(closest_time):
        return None
    else:
        MP_lag = lob[lob['datetime']==closest_time]['MP'].iloc[0]
        return MP_lag

lob['MP_lag_1m'] = lob['datetime'].apply(lambda x: get_closest_time(lob, x, pd.Timedelta('1min'))) 
lob['MP_lag_5m'] = lob['datetime'].apply(lambda x: get_closest_time(lob, x, pd.Timedelta('5min'))) 

# Calculate for 1 minute and 5 minutes
lob['MPC1'] = (lob['MP'] - lob['MP_lag_1m']) / lob['MP_lag_1m']
lob['MPC5'] = (lob['MP'] - lob['MP_lag_5m']) / lob['MP_lag_5m']

print(lob['MPC1'])
print(lob['MPC5'])
```

```python
# Skewness of Midpoint Price Change Rate

MPC1_mean = lob['MPC1'].mean()
MPC1_std = lob['MPC1'].std()
lob['MPC1_skew'] = ((lob['MPC1'] - MPC1_mean) / MPC1_std) ** 3
MPC1_skew = lob['MPC1_skew'].mean()

MPC5_mean = lob['MPC5'].mean()
MPC5_std = lob['MPC5'].std()
lob['MPC5_skew'] = ((lob['MPC5'] - MPC5_mean) / MPC5_std) ** 3
MPC5_skew = lob['MPC5_skew'].mean()

print(MPC1_skew)
print(MPC5_skew)
```

```python
# Maximum Midpoint Price Change

MPC1_max = lob['MPC1'].max()
MPC5_max = lob['MPC5'].max()

print(MPC1_max)
print(MPC5_max)
```

```python
# Mid-Price Basis

trade['datetime'] = pd.to_datetime(trade['datetime']).dt.floor('T')
trade_grouped = trade.groupby('datetime').agg({'price': 'mean'})
trade_grouped.rename(columns={'price':'avg_TP'}, inplace=True)

lob['datetime'] = pd.to_datetime(lob['datetime']).dt.floor('T')
lob_grouped = lob.groupby('datetime').agg({'MP': 'mean'})
# 计算每分钟的平均交易价格
lob_grouped['avg_MP'] = (lob_grouped['MP'] + lob_grouped['MP'].shift(1)) / 2

merged_trade_lob = pd.merge(trade_grouped, lob_grouped, on='datetime', how='left')
merged_trade_lob['MPB'] = merged_trade_lob['avg_TP'] - merged_trade_lob['avg_MP']

print(merged_trade_lob['MPB'])
```

```python
# Weighted Volume

def compute_weighted_volume(entries):
    levels = entries.split('|')
    volumes = [float(level.split(',')[1]) for level in levels]  # Consider all available levels
    total_levels = len(volumes)
    weights = [1 - (i - 1) / total_levels for i in range(1, total_levels + 1)]  
    weighted_volume = sum(w * v for w, v in zip(weights, volumes)) / sum(weights)
    return weighted_volume

lob['weighted_bid_volume'] = lob['bids'].apply(compute_weighted_volume)
lob['weighted_ask_volume'] = lob['asks'].apply(compute_weighted_volume)
```

```python
# Volume Order Imbalance

def delta_VWB(row, previous_row):
    if row['bid1'] < previous_row['bid1']:
        return 0
    elif row['bid1'] == previous_row['bid1']:
        return row['weighted_bid_volume'] - previous_row['weighted_bid_volume']
    else:
        return row['weighted_bid_volume']

def delta_VWA(row, previous_row):
    if row['ask1'] > previous_row['ask1']:
        return 0
    elif row['ask1'] == previous_row['ask1']:
        return row['weighted_ask_volume'] - previous_row['weighted_ask_volume']
    else:
        return row['weighted_ask_volume']

lob['delta_VWB'] = 0
lob['delta_VWA'] = 0

for i in range(1, len(lob)):
    lob.loc[i, 'delta_VWB'] = delta_VWB(lob.loc[i], lob.loc[i-1])
    lob.loc[i, 'delta_VWA'] = delta_VWA(lob.loc[i], lob.loc[i-1])

lob['VOI'] = lob['delta_VWB'] - lob['delta_VWA']

print(lob['VOI'])
```

```python
# Order Imbalance Ratio

lob['OIR'] = (lob['weighted_bid_volume'] - lob['weighted_ask_volume']) / \
             (lob['weighted_bid_volume'] + lob['weighted_ask_volume'])
print(lob['OIR'])
```

```python
# Step Order Imbalance Ratio

def extract_levels(entries):
    levels = entries.split('|')
    prices_volumes = [(float(level.split(',')[0]), float(level.split(',')[1])) for level in levels]
    return prices_volumes

lob['bid_levels'] = lob['bids'].apply(extract_levels)
lob['ask_levels'] = lob['asks'].apply(extract_levels)

def compute_soir(bid_levels, ask_levels):
    soirs = []
    num_levels = min(len(bid_levels), len(ask_levels))  # 获取买卖方的最小层级数
    for i in range(num_levels):
        V_B_i_t = bid_levels[i][1]  # 买方的量
        V_A_i_t = ask_levels[i][1]  # 卖方的量
        if V_B_i_t + V_A_i_t == 0:
            soirs.append(0)  # 避免除以零的情况
        else:
            soirs.append((V_B_i_t - V_A_i_t) / (V_B_i_t + V_A_i_t))
    return soirs, num_levels

lob['SOIR_levels'], lob['num_levels'] = zip(*lob.apply(lambda row: compute_soir(row['bid_levels'], row['ask_levels']), axis=1))

def compute_weighted_soir(soirs, num_levels):
    weights = [1 - (i-1)/num_levels for i in range(1, num_levels+1)]  # 动态计算权重
    weighted_soir = sum(w * soir for w, soir in zip(weights, soirs)) / sum(weights)
    return weighted_soir

lob['SOIR'] = lob.apply(lambda row: compute_weighted_soir(row['SOIR_levels'], row['num_levels']), axis=1)

print(lob['SOIR'])
```

```python
# Illiquidity

trade['Shortcut'] = 2 * (trade['high'] - trade['low']) - abs(trade['price'] - trade['open'])

trade = trade.dropna().reset_index(drop=True)

def calculate_illiq(df, p):
    df['ShortCut'] = df['Shortcut'].fillna(0)  
    df['Value'] = df['turnover']  
    grouped = df.resample(f'{p}T', on='datetime').agg({
        'ShortCut': 'sum',  
        'Value': 'sum'  
    }).reset_index()
    grouped['ILLIQ'] = grouped['ShortCut'] / grouped['Value']
    illiq_avg = grouped['ILLIQ'].mean()
    return illiq_avg
    
# Calculate for 5 minute and 15 minutes
illiq_5min = calculate_illiq(trade, 5)
print(illiq_5min)

illiq_15min = calculate_illiq(trade, 15)
print(illiq_15min)
```
