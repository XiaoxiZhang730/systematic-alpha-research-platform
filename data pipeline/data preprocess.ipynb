```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
import yfinance as yf

df = yf.download(
    "AAPL",
    start="2016-01-01",
    end="2025-12-31",
    interval="1d"
)
```

```python
df
```

```python
fig = plt.figure(figsize=(25,25))
i=1
for x in df.columns :
    plt.subplot(2,3,i)
    ax=sns.lineplot(df[x])
    ax.set(xlabel=None)
    plt.title(str(x), loc='center')
    plt.xlabel(None)
    plt.ylabel(None)
    i+=1
plt.show()
```
